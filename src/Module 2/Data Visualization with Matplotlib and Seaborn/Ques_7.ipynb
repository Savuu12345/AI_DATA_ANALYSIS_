```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Height': [160, 165, 170, 175, 180],
    'Weight': [55, 60, 65, 70, 75],
    'Size':   [10, 15, 13, 18, 12]
}
df = pd.DataFrame(data)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x='Height',
    y='Weight',
    size='Size',
    sizes=(50, 300),
    palette='deep',
    legend=False
)
plt.title("Height vs. Weight Scatter Plot with Size Encoding")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```
