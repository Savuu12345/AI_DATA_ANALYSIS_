```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', None],
    'Age': [28, 34, None, 41, 37],
    'City': ['New York', None, 'Boston', 'Chicago', 'Miami'],
    'Salary': [75000, 65000, 80000, None, 90000]
}
df = pd.DataFrame(data)
missing_data = df.isnull().sum()
plt.figure(figsize=(10, 6))
missing_data.plot(kind='bar')
plt.title('Missing Values by Column')
plt.ylabel('Count')
plt.xlabel('Columns')
plt.tight_layout()
plt.savefig('basic_plot.png')
plt.close()

print("Basic plot created.")
plt.figure(figsize=(12, 7))
ax = missing_data.plot(
    kind='bar',
    color='#5DA5DA',  
    edgecolor='black',
    width=0.7,
    alpha=0.8
)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title('Missing Values by Column', 
          fontsize=16, 
          fontweight='bold', 
          pad=20)
plt.ylabel('Number of Missing Values', fontsize=12)
plt.xlabel('DataFrame Columns', fontsize=12)
for i, v in enumerate(missing_data):
    ax.text(i, v + 0.1, str(v), ha='center', fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
ax.set_facecolor('#f8f9fa')
for spine in ax.spines.values():
    spine.set_edgecolor('#333333')
    spine.set_linewidth(1.5)
plt.figtext(0.02, 0.02, 'Generated with Matplotlib', 
            fontsize=8, style='italic', ha='left')

avg_missing = missing_data.mean()
plt.axhline(y=avg_missing, color='red', linestyle='-.', alpha=0.7, 
            label=f'Average ({avg_missing:.2f})')
plt.legend()
plt.tight_layout()
plt.savefig('customized_plot.png', dpi=300, bbox_inches='tight')
print("Customized plot created.")
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(12, 7))
colors = ['#6495ED', '#7EB6FF', '#4169E1', '#1E90FF', '#87CEFA']
ax = missing_data.plot(
    kind='bar',
    ax=ax,
    color=colors[:len(missing_data)],
    edgecolor='black',
    width=0.65
)
for i, v in enumerate(missing_data):
    ax.text(i, v + 0.1, str(v), ha='center', fontweight='bold')
    percentage = (v / len(df)) * 100
    ax.text(i, v/2, f'{percentage:.1f}%', ha='center', 
            color='white', fontweight='bold')
plt.title('Data Completeness Analysis', 
          fontsize=18, 
          fontweight='bold', 
          pad=20)
plt.ylabel('Missing Value Count', fontsize=14)
plt.xlabel('Data Categories', fontsize=14)
plt.figtext(0.5, 0.92, f'Total Records: {len(df)}, Analyzed on: {pd.Timestamp.now().strftime("%Y-%m-%d")}',
            ha='center', fontsize=10)
plt.xticks(rotation=30, ha='right', fontsize=12)
plt.yticks(fontsize=12)
summary_text = f"""
Dataset Summary:
 Total records: {len(df)}
 Columns with missing data: {sum(missing_data > 0)}
 Total missing values: {missing_data.sum()}
 Missing data ratio: {missing_data.sum() / (len(df) * len(df.columns)):.2%}
"""
props = dict(boxstyle='round', facecolor='white', alpha=0.7)
plt.figtext(0.15, 0.15, summary_text, fontsize=9,
            verticalalignment='top', bbox=props)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
plt.tight_layout()
plt.savefig('professional_plot.png', dpi=300, bbox_inches='tight')

print("Professional themed plot created.")
print("\nAvailable font families:")
font_list = sorted([f.name for f in font_manager.fontManager.ttflist])[:5]
print(font_list, "... and more")
print("\nMatplotlib customization complete!")
```
