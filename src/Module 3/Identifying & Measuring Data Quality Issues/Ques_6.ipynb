```python
# Question: Evaluating Data Distribution
# Description: Analyze the distribution of a numeric column using histograms and descriptive statistics.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
def analyzedistribution(data,column_name):
    if column_name not in data.columns:
        print(f"Error:Column'{column_name}'notfoundintheDataFrame.")
        return
    if not pd.api.types.is_numeric_dtype(data[column_name]):
        print(f"Error:Column'{column_name}'isnotnumeric.")
        return
    print(f"DescriptiveStatisticsfor'{column_name}':")
    print(data[column_name].describe())
    plt.figure(figsize=(10,6))
    sns.histplot(data[column_name],kde=True)
    plt.title(f'Distributionof{column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.show()
if __name__=='__main__':
    sample_data=pd.DataFrame({'Value':[10,12,12,13,12,11,14,13,15,10,10,10,100],'Category':['A','B','A','C','B','A','C','B','A','A','B','A','C']})
    analyzedistribution(sample_data,'Value')
    analyzedistribution(sample_data,'Category')
    analyzedistribution(sample_data,'NonExistentColumn')
```
