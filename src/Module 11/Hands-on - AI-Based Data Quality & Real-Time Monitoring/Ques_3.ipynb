## Isolation Forest for Anomaly Detection
**Objective**: Understand and apply the Isolation Forest algorithm to identify anomalies in datasets.

### Task: Anomaly Detection in Sensor Data
**Steps**:
1. Load Dataset
2. Feature Selection
3. Isolation Forest Implementation
4. Plot Results

```python

import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
normal_data=np.random.normal(loc=[50,100],scale=[5,10],size=(950,2))
anomalies=np.random.normal(loc=[10,150],scale=[2,5],size=(50,2))
sensor_data=np.vstack((normal_data,anomalies))
df=pd.DataFrame(sensor_data,columns=['temperature','pressure'])
model=IsolationForest(n_estimators=100,contamination=0.05,random_state=42)
df['anomaly_score']=model.fit_predict(df[['temperature','pressure']])
anomalies_detected=df[df['anomaly_score']==-1]
normal_data_points=df[df['anomaly_score']==1]
plt.figure(figsize=(12,8))
plt.scatter(normal_data_points['temperature'],normal_data_points['pressure'],c='blue',label='Normal Readings',alpha=0.6)
plt.scatter(anomalies_detected['temperature'],anomalies_detected['pressure'],c='red',label='Anomalous Readings',s=100,edgecolors='k')
plt.title('Anomaly Detection in Sensor Data using Isolation Forest')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.show()
```
